```python
from run import Model
from mmengine import Config
import torch
from matplotlib import pyplot as plt
import numpy as np
import torchvision
```

```python
model = Model(Config.fromfile('07-14-24/16:24:09-[gen-models]-[ddim-flower-large]/config.py'))
model.load_state_dict(torch.load('07-14-24/16:24:09-[gen-models]-[ddim-flower-large]/last.ckpt')['state_dict'])
model.to('cuda')
model.eval()
```
